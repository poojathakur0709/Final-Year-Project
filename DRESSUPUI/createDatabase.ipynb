```python
def ConvertToBinaryData(filename):
    #Convert digital data to binary format
    with open(filename, 'rb') as file:
        blobData = file.read()
    return blobData
```

```python
def insertUser(username,Password ,DOB ,Gender,Skintone):
    conn = sqlite3.connect('DressUP.db')
    cursor = conn.cursor()
    cursor.execute('''Select * from User where Email = ?''',[username])
    length = len(cursor.fetchall())
    if length >0:
        conn.commit()
        cursor.close()
        return False
    else:
        data_typle = (username ,Password ,DOB ,Gender,Skintone)
        cursor.execute("INSERT INTO User(Email ,Password ,DOB ,Gender,Skintone) VALUES(?,?,?,?,?)", data_typle)
        conn.commit()
        cursor.close()
        return True
```

```python
def CreateDatabase():
    #Connect to the database
    conn = sqlite3.connect('DressUP.db')
    #Create a cursor object
    cursor = conn.cursor()
    cursor.execute("DROP TABLE IF EXISTS User")
    cursor.execute("DROP TABLE IF EXISTS Storage")
    cursor.execute("DROP TABLE IF EXISTS UpperStorage")
    cursor.execute("DROP TABLE IF EXISTS footStorage")
    cursor.execute("DROP TABLE IF EXISTS lowerStorage")
    
    cursor.execute('''CREATE TABLE User
(
    Email TEXT, 
    Password TEXT, 
    DOB TEXT,
    Gender TEXT, 
    Skintone TEXT,
    CONSTRAINT users_pk PRIMARY KEY (Email)
);''') 
    conn.commit()
    cursor.execute('''CREATE TABLE UpperStorage
(
    ID INTEGER PRIMARY KEY AUTOINCREMENT UNIQUE NOT NULL,
    Type TEXT,
    Gender TEXT,
    Color TEXT,
    Season TEXT,
    Usage TEXT,
    Image BLOB,
    Email TEXT,
    h TEXT,
    s TEXT,
    v TEXT,
    CONSTRAINT fk_email
    FOREIGN KEY (Email)
    REFERENCES User(Email)
);''') 
    conn.commit()
    cursor.execute('''CREATE TABLE footStorage
(
    ID INTEGER PRIMARY KEY AUTOINCREMENT UNIQUE NOT NULL,
    Type TEXT,
    Gender TEXT,
    Color TEXT,
    Season TEXT,
    Usage TEXT,
    Image BLOB,
    Email TEXT,
    h TEXT,
    s TEXT,
    v TEXT,
    CONSTRAINT fk_email
    FOREIGN KEY (Email)
    REFERENCES User(Email)
);''') 
    conn.commit()
    cursor.execute('''CREATE TABLE lowerStorage
(
    ID INTEGER PRIMARY KEY AUTOINCREMENT UNIQUE NOT NULL,
    Type TEXT,
    Gender TEXT,
    Color TEXT,
    Season TEXT,
    Usage TEXT,
    Image BLOB,
    Email TEXT,
    h TEXT,
    s TEXT,
    v TEXT,
    CONSTRAINT fk_email
    FOREIGN KEY (Email)
    REFERENCES User(Email)
);''') 
    conn.commit()
    cursor.close()
```

```python
def convert_data(data, file_name):

    # Convert binary format to
    # images or files data
    with open(file_name, 'wb') as file:
        file.write(data)
```
